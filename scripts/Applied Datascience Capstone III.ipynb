```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
df_list = pd.read_html(url)
```

```python
dftable = df_list[0]
```

```python
dftable
```

```python
dftable2 = dftable.set_index('Borough').drop('Not assigned', axis=0)
dftable2
```

```python
dftable2 = dftable2.reset_index()
```

```python
dftable2
```

```python
dftable3 = pd.DataFrame(dftable2, columns=['Postal Code', 'Borough', 'Neighbourhood'])
dftable3
```

```python
url2 = "https://cocl.us/Geospatial_data"
geo_data = pd.read_csv(url2)
geo_data
```

```python
geo_data = geo_data.set_index('Postal Code')
dftable4 = dftable3.set_index('Postal Code')
result = dftable4.join(geo_data)
result = result.reset_index()
result
```

```python
result2 = result[result['Borough'].str.contains('Toronto',regex=False)]
result2
```

```python
# create map of New York using latitude and longitude values
map_newyork = folium.Map(location=[43.651070,-79.347015], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(result['Latitude'], result['Longitude'], result['Borough'], result['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```
